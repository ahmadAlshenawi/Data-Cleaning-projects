```python
new_columns_names = []

for i in data.columns :
    new_columns_names.append(i.capitalize())
```

```python
data["Show_id"].unique()
# we need to transform the show_id column to be a string type instead of int type because it is an id and we don't need to do any calculations on it
```

```python
data["Show_id"] = data["Show_id"].apply(lambda x : x.split("s")[1])
```

```python
data
```

```python
type(data["Show_id"][0])
```

```python
data["Show_id"] = data["Show_id"].astype(int)
```

```python
data.info()   
```

```python
data["Date_added"] = data["Date_added"].apply(lambda x : x.replace("/","-"))
```
